```python
preg = preg.iloc[:, 0:10]
preg.head()
```

```python
preg.columns = [
"Name",
"Age",
"Gravida",
"TetanusDose",
"GestationalWeeks",
"Weight",
"Height",
"BloodPressure",
"Anemia",
"Jaundice"
]

preg.head()
```

```python
preg = preg.drop(index=0)
preg.head()
```

```python
preg["GestationalWeeks"] = preg["GestationalWeeks"].str.extract("(\d+)").astype(float)
preg["Weight"] = preg["Weight"].str.extract("(\d+)").astype(float)

preg[["SystolicBP","DiastolicBP"]] = preg["BloodPressure"].str.split("/", expand=True)

preg["SystolicBP"] = preg["SystolicBP"].astype(float)
preg["DiastolicBP"] = preg["DiastolicBP"].astype(float)

preg.head()
```
